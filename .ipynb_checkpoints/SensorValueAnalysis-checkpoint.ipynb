```python
## Sonar Data is viewed for sample here

import numpy as np
import matplotlib.pyplot as plt
%matplotlib inline 

#loading the sonar measurements from the np bin file .npy
SonarMeasure = np.load("/home/adithya/DAQ-Sensor/DataStored/measurement0025.npy")
#printing the data first measurements
print(SonarMeasure[:, :, 0])
#shape of the data
print(SonarMeasure.shape)

Left = SonarMeasure[:,1,0]
Right = SonarMeasure[:,0,0]


#plotting the data 
fig = plt.figure()
#w = [[pow(b[0],2) + 0.0000000001 ,pow(b[1],2) + 0.000000000001] for b in SonarMeasure[:,:,1]]
#print(w)
bx = fig.add_subplot(121)
ax = fig.add_subplot(122)
bx.plot(SonarMeasure[:,:,1])
plt.show()

fig2 = plt.figure()
bx = fig2.add_subplot(121)
ax = fig2.add_subplot(122)
bx.plot(Right, color = 'red')
ax.plot(Left, color ="blue")
plt.show()

diff = Left - Right

RightCutoff = np.where(Right>2, Right , 0)
LeftCutoff = np.where(Left>2, Left , 0)

fig3 = plt.figure()
ax = fig3.add_subplot(121)
ax.plot(LeftCutoff, color = 'yellow')
bx = fig3.add_subplot(122)
bx.plot(RightCutoff, color = 'green')
plt.show()

WindowSize =10 
LeftW = np.array([sum(pow(Left[i:i+WindowSize],2)) for i in range(len(Left))])
RightW = np.array([sum(pow(Right[i:i+WindowSize],2)) for i in range(len(Right))])
print(LeftW)
LeftW = 10*np.log10(RightCutoff)
RightW = 10*np.log10(LeftCutoff)
fig4 = plt.figure()
ax = fig4.add_subplot(121)
bx = fig4.add_subplot(122)
ax.plot(LeftW, color = 'orange')
bx.plot(RightW, color = 'skyblue')
plt.show()



Diff_power = (LeftW - RightW)
fig3 = plt.figure()
ax = fig3.add_subplot(121)
ax.plot(Diff_power, color = 'yellow')
plt.show()
```
